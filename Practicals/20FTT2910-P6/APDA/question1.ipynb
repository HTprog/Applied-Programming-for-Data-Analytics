PART A K-MEANS CLUSTERING

```python
X['Gender'].replace(to_replace={'Male':0,'Female':1},inplace=True)
```

```python

from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

inertia_list = []
for num_clusters in range(1, 15):
    kmeans_model = KMeans(n_clusters=num_clusters, init="k-means++")
    kmeans_model.fit(X)
    inertia_list.append(kmeans_model.inertia_)
    
# plot the inertia curve
plt.plot(range(1,15),inertia_list)
plt.scatter(range(1,15),inertia_list)
plt.xlabel("Number of Clusters", size=14)
plt.ylabel("Inertia Value", size=14)
plt.title("Different Inertia Values for Different Number of Clusters", size=14,y=1.03)
plt.text(x=-6,y=0,s='Best cluster is either 4 or 5',color='g')
plt.show()
```

```python
#original labels
kmeans = KMeans()
kmeans.fit(X)
assignments = kmeans.labels_
mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignments)
plt.title('Original label')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
#cluster=1
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)
assignments = kmeans.labels_
mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignments)
plt.title('Cluster=1')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
fig, axes = plt.subplots(1,4, figsize=(40, 10),sharey=True)

#cluster=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
assignments = kmeans.labels_
axes[0].set_title('n_clusters=2',size=20)
axes[0].set_xlabel('Annual Income (k$)')
axes[0].set_ylabel('Spending Score (1-100)')
mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignments, ax=axes[0])

#cluster=3
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
assignments = kmeans.labels_
axes[1].set_title('n_clusters=3',size=20)
axes[1].set_xlabel('Annual Income (k$)')

mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignments, ax=axes[1])

#cluster=4
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
assignments = kmeans.labels_
axes[2].set_title('n_clusters=4',size=20)
axes[2].set_xlabel('Annual Income (k$)')

mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignments, ax=axes[2])

# #cluster=5
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
assignments = kmeans.labels_
axes[3].set_title('n_clusters=5',size=20)
axes[3].set_xlabel('Annual Income (k$)')

mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4], assignments, ax=axes[3])
plt.legend(['Cluster 1','Cluster 2','Cluster 3','Cluster 4','Cluster 5'])
plt.show()
```

PART B AGGLOMERATIVE CLUSTERING

```python
#original labels
from sklearn.cluster import AgglomerativeClustering
agg=AgglomerativeClustering()
assignment=agg.fit_predict(X)
plt.title('Original Labels')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignment)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
fig, axes = plt.subplots(1,4, figsize=(40, 10),sharey=True)


agg=AgglomerativeClustering(n_clusters=1)
assignment=agg.fit_predict(X)
axes[0].set_title('K=1')
axes[0].set_xlabel('Annual Income (k$)')
axes[0].set_ylabel('Spending Score (1-100)')
mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignment,ax=axes[0])

agg=AgglomerativeClustering(n_clusters=2)
assignment=agg.fit_predict(X)
axes[1].set_title('K=2')
axes[1].set_xlabel('Annual Income (k$)')

mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignment,ax=axes[1])

agg=AgglomerativeClustering(n_clusters=3)
assignment=agg.fit_predict(X)
axes[2].set_title('K=3')
axes[2].set_xlabel('Annual Income (k$)')

mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignment,ax=axes[2])

agg=AgglomerativeClustering(n_clusters=4)
assignment=agg.fit_predict(X)
axes[3].set_title('K=4')
axes[3].set_xlabel('Annual Income (k$)')

mglearn.discrete_scatter(X.iloc[:, 3],X.iloc[:, 4],assignment,ax=axes[3])

plt.legend(['Cluster 1','Cluster 2','Cluster 3','Cluster 4'])
plt.show()
```
