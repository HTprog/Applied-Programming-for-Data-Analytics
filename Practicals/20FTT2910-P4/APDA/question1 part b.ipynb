```python
sns.pairplot(data=df)
plt.show()
#This visualization is a pairplot of the 'Sales.csv' dataset. It provides a general overview of the pairwise relationships of 
# every numerical attributes within the dataset.
```

```python
plt.scatter('Customer_Age','Revenue',data=df)
plt.xlabel('Customer Age')
plt.ylabel('Revenue')
plt.title('Customer Age against Revenue')
plt.show()
#This visualization shows a scatterplot of Customer Age against Revenue. It shows there are two outliers at age 50 and it can be seen that revenue suffers a drastic drop when customer age is above 60.
```

```python
sns.countplot(x='Product_Category',data=df)
plt.show()
#According to this countplot of each product category it can be seen that 
# the most frequent category is 'Accessories' while 'Bikes' comes second and finally 'Clothing' in third.
```

```python
g=sns.catplot(x='Sub_Category',data=df,col='Product_Category',kind='count')
plt.xlim(xmin=1)
g.set_xticklabels(rotation=90)
plt.show()
#This visualization shows multiple counts of each sub category from each product category. 
# It shows the most popular category of accessories are 'Tires and Tubes',
# the most popular category of Clothing are 'Jerseys' and the most popular category of bikes are 'Road Bikes'
```

```python
# label=list(df['Age_Group'].unique())
# h=label.pop(2)
# label.insert(1,h)
plt.figure(figsize=(10,5))
p=sns.histplot(x='Age_Group',data=df)
plt.show()
#This visualization shows a histogram of 'Age_Group' as its bins. According to the graph, 
# Adults(35-64) are the most numerous type of customer while Young Adults(25-34) are the second most numerous, 
# Youth(below 25) as third and Seniors(above 64) as fourth.
```

```python
plt.figure(figsize=(10,15))
sns.barplot(x='Profit',y='State',data=df)
plt.show()
# This visualization shows a barplot of the profits of each state. 
# It is shown that the state of Arizona has made the. most profit out of all states
```
