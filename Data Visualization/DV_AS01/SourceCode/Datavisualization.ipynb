```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
df=pd.read_csv('Train-Set(prepared).csv')
df
```

```python
sns.countplot(y='ProductType',data=df,orient='h')
plt.xticks(range(0,1201,100))
plt.title('Countplot of each product type')
plt.show()
```

```python
sns.boxplot(y='ProductType',x='MRP',data=df,orient='h')
plt.title('Average MRP of each ProductType')
plt.show()
```

```python
sns.countplot(x='EstablishmentYear',hue='FatContent',data=df)
plt.title('FatContent in each establishment')
plt.show()
```

```python
avg_pv=df.groupby('ProductType').mean()
idx=pd.IndexSlice
dat=avg_pv.loc[idx[:],idx['ProductVisibility']]
plt.barh(y=dat.index,width=dat,data=dat)
plt.title('Average ProductVisibility of each ProductType')
plt.xlabel('Average ProductVisibility')
plt.ylabel('ProductType')
plt.show()
```

```python
sns.scatterplot(x='ProductVisibility',y='OutletSales',data=df)
plt.title('Change in OutputSales against ProductVisibility scatterplot')
plt.show()
```

```python

sum_sales=df.groupby('OutletType').sum()
idx=pd.IndexSlice
fig, ax = plt.subplots(figsize = (10,10),facecolor='lightgrey')
sum=sum_sales.loc[idx[:],idx['OutletSales']]
plt.title('Total sales achieved by each OutletType')
plt.pie(x=sum,labels=sum.index,autopct='%1.0f%%')
plt.show()
```
