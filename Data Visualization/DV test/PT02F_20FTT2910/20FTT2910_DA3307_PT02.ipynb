```python
#Question 1A
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

array1=['Jan','Feb','March','April','May','Jan','Feb','March','April','May']
array2=[4,7,5,9,3,2,1,6,8,10]

plt.scatter(x=array1,y=array2)
plt.show()
```

```python
#Question 1B
df=pd.read_csv('PT02_1.csv')

Malepil=df['Male Pilgrims'].sum()
Femalepil=df['Female Pilgrims'].sum()

label=['Male','Female']
genders=[Malepil,Femalepil]

plt.pie(genders,labels=label,autopct='%.2f %%')
plt.title('Hajj pilgrims by gender 2005 to 2019')
plt.show()
```

```python
#Question 1C
df=pd.read_csv('PT02_1.csv')
filtdf=df.loc[(df['Year']==2018) | (df['Year']==2019)].sum()

label=['Male','Female']
values=[filtdf[2],filtdf[3]]

plt.pie(values,labels=label,autopct='%.f %%')
plt.title('Hajj pilgrims by gender in 2018 and 2019')
plt.show()
```

```python
#Question 2A
df2=pd.read_csv('PT02_3.csv')

districts=['Brunei/Muara','Tutong','Belait','Temburong']
values=[df2.iloc[:,1],df2.iloc[:,2],df2.iloc[:,3],df2.iloc[:,4]]

plt.title('Distribution of chicken meat production in all districts')
plt.boxplot(values,labels=districts)
plt.show()


#Question 2A Part III
#1.Brunei/Muara produces the most chicken meat on average most likely due to the district having the highest population of Bruneians 
#which in turn requires production to be higher than other districts to meet their needs.

#2.Temburong produces almost no chicken meat indicating little demand of chicken meat within that district.
```

```python
#Question 2B
df2=pd.read_csv('PT02_3.csv')
df2['Total(Month)']=df2['Brunei/Muara']+df2['Tutong']+df2['Belait']+df2['Temburong']

plt.figure(figsize=(14,6))
plt.bar(x=df2['Month'],height=df2['Total(Month)'])

plt.title('Chicken meat production')
plt.xlabel('Months')
plt.ylabel('Production in KG')

plt.show()
```

```python
#Question 3A
df2=pd.read_csv('PT02_3.csv')
plt.figure(figsize=(15,7))
for chicken in df2:
    if chicken !='Month':
        plt.plot(df2['Month'],df2[chicken],label=chicken)

months={1:'January',2:'Febuary',3:'March',4:'Ápril',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
plt.xticks(df2['Month'],labels=months)
plt.title('Chicken production in each district across the entire year ')
plt.legend()
plt.show()



#Question 3A part VI
#Ans=The month that produces the most chicken meat in Brunei/Muara is December while for Tutong it is September.


#Question 3A part VII
#Ans=Belait and Brunei/Muara suffered a decrease in chicken production while Tutong experienced an increase in chicken production.
```
