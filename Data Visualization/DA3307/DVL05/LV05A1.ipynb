```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[2,4,6]

a=[1,2,3,5]
b=[3,7,10,4]


plt.figure(figsize=(6,6), dpi=150)

plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,2,4,6,8,10])

plt.plot(x,y)
plt.plot(a,b)

plt.savefig('myfirstgraph.png',dpi=80)
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]

#plt.plot(x,y,label='line biru',color='green',marker='.',markersize=10,markeredgecolor='blue')
plt.plot(x,y,label='line biru',color='green',linestyle='--', linewidth=3)
         
plt.title('Our first graph',y=1.02,fontdict={'fontsize':25})

plt.xlabel('x label',loc="center",fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.ylabel('y label',fontdict={'fontsize':20})

plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]

plt.title('Our first graph')

plt.xlabel('x label')
plt.ylabel('y label')

plt.plot(x[:2],y[:2])
plt.plot(x[1:],y[1:],'b--')


plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
label=['A','B','C']

value=[1,2,3]

mybar=plt.bar(label,value)

plt.xticks(['A','B','C','D'])
plt.yticks([1,3,5,7])

mybar[0].set_hatch('/')
mybar[2].set_hatch('*')
plt.show()
```
