```python
#QUESTION 2 PART A
import pandas as pd
df=pd.read_csv('iris_pt.csv')
df.head(10)
```

```python
#QUESTION 2 PART B SUB-QUESTION i
iris_labels=df.pop('Species')
iris_labels
```

```python
#QUESTION 2 PART B SUB-QUESTION ii
df.drop('Id',axis=1,inplace=True)
df
```

```python
#QUESTION 2 PART B SUB-QUESTION iii,iv,v
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df, iris_labels, stratify=iris_labels,test_size=0.30, random_state=3)
```

```python
#QUESTION 2 PART C SUB QUESTION i,ii
from sklearn.tree import DecisionTreeClassifier 
tree=DecisionTreeClassifier()
tree.fit(X_train , y_train)
print ("Accuracy on training set: {:.2f}".format(tree.score(X_train , y_train)))
print ("Accuracy on test set: {:.2f}".format(tree.score(X_test , y_test)))
```

```python
#QUESTION 2 PART D SUB QUESTION i,ii
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
print("Accuracy on training set: {:.2f}".format(svc.score(X_train, y_train)))
print("Accuracy on test set: {:.2f}".format(svc.score(X_test, y_test)))
```
