```python
#QUESTION 1 PART A
import pandas as pd
from numpy import nan as NaN
df=pd.read_csv('cars_formal_pt.csv')
df.head(10)
```

```python
df.isna().sum()#before replacement of 0 values
```

```python
#QUESTION 1 PART B
df.replace([" ",0],NaN,inplace=True)
df.isna().sum()
```

```python
#QUESTION 1 PART C SUB-QUESTION i
dfdropped=df.dropna(axis=1,thresh=2)
dfdropped.isna().sum()
```

```python
#QUESTION 1 PART C SUB-QUESTION ii,iii
dffilled=dfdropped.fillna({'price':dfdropped['price'].mean(),'distance_travelled(kms)':dfdropped['distance_travelled(kms)'].mean(),'year':dfdropped['year'].mode()[0],'full_model_name':df['full_model_name'].mode()[0],'fuel_type':df['fuel_type'].mode()[0],'city':df['city'].mode()[0]},axis=0)
dffilled.isna().sum()
```

```python
#QUESTION 1 PART D
Brands={'Corolla':'Toyota','Fortuner':'Toyota','Innova':'Toyota','Etios':'Toyota','Verna':'Hyundai','Creta':'Hyundai','i20':'Hyundai','i10':'Hyundai'}
dffilled['Brand']=dffilled['model_name'].map(Brands)
dffilled
```

```python
#QUESTION 1 PART E SUB-QUESTION i
dffilled['Brand'].value_counts()
```

```python
print('oldest year:')
dffilled['year'].describe()['min']
```

```python
print('latest year:')
dffilled['year'].describe()['max']
```
