```python
import pandas as pd
import numpy as np
df=pd.read_csv('weatherAUS.csv')
df
```

```python
print('Duplicated data: '+ str(df.duplicated().sum()))
print("-------------------")
print('Missing values: \n'+ str(df.isnull().sum()))
```

```python
Columns=['MinTemp','MaxTemp','Evaporation','Sunshine','WindGustSpeed','WindSpeed9am','WindSpeed3pm',
'Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm']
Columns2=['WindGustDir','WindDir9am','WindDir3pm']

df.dropna(subset=['RainToday','RainTomorrow'],inplace=True)

for items in Columns:
    df.fillna({items:df[items].mean()},inplace=True)

for items in Columns2:
    df.fillna({items:df[items].mode()[0]},inplace=True)
```

```python
print('Duplicated data: '+ str(df.duplicated().sum()))
print("-------------------")
print('Missing values: \n'+ str(df.isnull().sum()))
```

```python
df.to_csv('weatherAUS(Prepared)1.csv',index=False)
```
